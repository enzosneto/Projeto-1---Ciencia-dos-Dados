pergunta: Após a entrada na união europeia, houve aumento na exportação da hungria comparado aos outros paises presentes na uniao europeia?
Após a entrada na uniao europeia a Hungria teve um aumento de aproximadamente 10 Bilhoes de Dolares so no primeiro ano. Esse aumento é algo padrao para os paises que entraram na uniao ou foi um caso a parte?

Hipótese: Todos paises que entraram na uniao europeia tem um aumento ridiculamente grande no quesito de exportacao em funcao do PIB.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import os
print('Esperamos trabalhar no diretório')
print(os.getcwd())

dados = pd.read_excel('exports_percent_of_gdp.xlsx')

```

```python
index=dados.set_index("country")
lista_paises=["Hungary"]
index.loc[:, "Hun"]=0
index.loc[lista_paises, "Hun"]=1
dados_Hun=index.loc[lista_paises]
dados_Hun

transpor=index.transpose()
transpor_quant=transpor[:-1]
ue_anos=transpor_quant.loc[(transpor_quant.index>1999) & (transpor_quant.index<2009),:].transpose()

lista_ue=["Austria","Belgium","Bulgaria","Croatia","Cyprus","Czechia","Denmark","Estonia","Finland","France","Germany","Greece","Hungary","Ireland","Italy","Latvia","Lithuania","Luxembourg","Malta","Netherlands","Poland","Portugal","Romania","Slovakia","Slovenia","Spain","Sweden","United Kingdom"]
dados_uniao_europeia=ue_anos.loc[lista_ue]
dados_uniao_europeia
```

# Anos antes e depois da adesão da Hungria na UE 

```python
transpor=dados_Hun.transpose()
transpor_quant=transpor[:-1]
hun_anos=transpor_quant.loc[(transpor_quant.index>2000) & (transpor_quant.index<2009),:].transpose()
hun_anos

```
